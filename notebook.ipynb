# Gross Domestic Product Data
This dataset consists of the yearly gross domestic product (GDP) in current USD of countries and regions worldwide since the 1960s.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
gdp_data = pd.read_csv("gdp_data.csv", index_col=None)
print(gdp_data.head)
gdp_data.head(10)
```

```python
#See the wholistic shape of the data
gdp_data.shape  #rows = 11507 and columns = 4
```

```python
gdp_data.sort_values(by="Year", ascending=True)
```

```python
#Get sum of unique values in each columns
gdp_data.nunique()
```

```python
#Exploring the unique values in column 'Year'
unique_years = sorted(gdp_data['Year'].unique())
np.array(unique_years)
```

```python
#Exploring unique values in column 'Country Name'
unique_countries = gdp_data['Country Name'].unique()
np.array(unique_countries)
```

## Note:  
The beginning values in unique_countries seems like some values other than real-world countries

```python
# Define the range of country names to be removed
start_country = 'Arab World'
end_country = 'Upper middle income'

# Find the indices of the start and end countries
start_index = list(unique_countries).index(start_country)
end_index = list(unique_countries).index(end_country)

# Remove the countries in the specified range
filtered_countries = [country for i, country in enumerate(unique_countries) if i < start_index or i > end_index]

filtered_countries
```

## Again!
The list still contains a non-country value 'World'

```python
# Remove 'World' from filtered_countries
filtered_countries = [country for country in filtered_countries if country != 'World']

np.array(filtered_countries)
```

```python
# Update gdp_data dataframe with only countries in filtered_countries
gdp_data = gdp_data[gdp_data['Country Name'].isin(filtered_countries)]

gdp_data
```

```python
india_gdp_data = gdp_data[gdp_data['Country Code'] == 'IND']
india_gdp_data.head()
```

```python
# Assuming india_gdp_data and world_gdp_data are DataFrames
plt.plot(india_gdp_data['Year'], india_gdp_data['Value'], color='red')
plt.title('India GDP Past Decades')
plt.xlabel('Year')
plt.ylabel('GDP Value')
plt.legend(['India GDP'])
plt.show()
```
